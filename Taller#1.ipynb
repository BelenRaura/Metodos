# Tarea Graficar
$y(x)=x^3 -7 x^2 +4x +12$

|

```python
def ecuacion(x:int)->int:
    return x**3 - 7*x**2 +  4*x+12
    
```

```python
import numpy as np
X,step= np.linspace(-10,10,retstep=True)
X
```

```python
Y =[]
for x in X:
    y=ecuacion(x)
   # X.append(x)
    Y.append(y)

    print(f'X= {x:.2f} \t Y= {y:.2f}')
```

```python
from matplotlib import pyplot as plt

plt.plot(X, Y, marker="o", color="red")
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title("Funcion Cubica")
plt.legend(['$y=x^3 -7 x^2 +4x +12$'])


plt.ion()

plt.show()
```
