```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

def f(x):
    return x**3 - 3*x**2 + x - 1

def df(x):
    return 3*x**2 - 6*x + 1

def newton_method(f, df, x0, tol=1e-6, max_iter=100):
   

    x = x0
    points = [x]
    for _ in range(max_iter):
        try:
            x_new = x - f(x) / df(x)
        except ZeroDivisionError:
            print(f"Divisin por cero en x={x}. La derivada se anula.")
            return None, points
        points.append(x_new)
        if abs(x_new - x) < tol:
            break
        x = x_new
    return x, points

# Valores iniciales
x0s = [-1, 0.5, 2]

# Grfica de la funcin
x = np.linspace(-2, 3, 100)
y = f(x)
plt.plot(x, y, label='f(x)')
plt.grid(True)

# Encontrar races y graficar la trayectoria
for x0 in x0s:
    root, points = newton_method(f, df, x0)
    if root is not None:
        plt.plot(points, f(np.array(points)), 'o-', label=f'x0={x0}')
        print(f"Raz encontrada para x0={x0}: {root}")
    else:
        print(f"No se encontr raz para x0={x0}")

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Mtodo de Newton')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

def f(x, points):
    y = x**3 - 3*x**2 + x - 1
    points.append((x, y))
    return y

def df(x):
    return 3*x**2 - 6*x + 1

# Valor inicial
x0 = 100

# Lista para almacenar los puntos visitados
points = []

# Aplicamos el mtodo de Newton y almacenamos los puntos
root = newton(f, x0, args=(points,))

print("La raz encontrada es:", root)

# Extraemos las coordenadas x e y de los puntos
x_values, y_values = zip(*points)

# Creamos la figura y los ejes
plt.figure()

# Graficamos la funcin
x = np.linspace(-2, 3, 100)
y = f(x, [])  # Llamamos a f sin modificar points para obtener los valores de y
plt.plot(x, y, label='f(x)')

# Graficamos los puntos visitados
plt.scatter(x_values, y_values, color='red', label='Puntos visitados')

# Personalizamos la grfica
plt.xlabel('x')
plt.ylabel('y')
plt.title('Mtodo de Newton')
plt.grid(True)
plt.legend()

# Mostramos la grfica
plt.show()
```
